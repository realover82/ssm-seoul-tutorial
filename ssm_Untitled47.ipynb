```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
import os

# 데이터 로드 및 전처리 함수
def load_data(directory):
    images = []
    labels = []
    for class_name in os.listdir(directory):
        class_dir = os.path.join(directory, class_name)
        if os.path.isdir(class_dir):
            for filename in os.listdir(class_dir):
                if filename.endswith('.jpg') or filename.endswith('.png'):
                    img_path = os.path.join(class_dir, filename)
                    img = Image.open(img_path)
                    img = img.resize((224, 224))  # 이미지 크기 조절
                    img = img.convert('RGB')  # 이미지를 RGB로 변환
                    img = np.array(img)  # 이미지를 배열로 변환
                    images.append(img)
                    labels.append(class_name)
    return np.array(images), np.array(labels)

# 데이터 경로 설정
positive_data_dir = "/content/drive/MyDrive/deeplearning/remocon/ssm-original4"  # 양품 사진 데이터 폴더 경로
negative_data_dir = "/content/drive/MyDrive/deeplearning/remocon/ssm-scratch"  # 스크래치 불량 데이터 폴더 경로

# 데이터 로드
X_positive, y_positive = load_data(positive_data_dir)
X_negative, y_negative = load_data(negative_data_dir)

# 데이터 결합 및 레이블 지정
X = np.concatenate((X_positive, X_negative), axis=0)
y = np.concatenate((np.zeros(len(X_positive)), np.ones(len(X_negative))), axis=0)

# 데이터 섞기
random_indices = np.random.permutation(len(X))
X = X[random_indices]
y = y[random_indices]

# 데이터 분할 (학습 데이터와 테스트 데이터로 분할)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Random Forest 모델 학습
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train.reshape(len(X_train), -1), y_train)

# 모델 평가
y_pred = clf.predict(X_test.reshape(len(X_test), -1))

# 평가 지표 계산
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# 결과 출력
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import os
import numpy as np
from PIL import Image

# 데이터 폴더 경로
data_directory = "/content/drive/MyDrive/deeplearning/remocon/test03-03-01/button01"


# 이미지와 레이블을 담을 빈 리스트
images = []
labels = []

# 주어진 디렉토리 내의 각 클래스(폴더)를 반복
for class_name in os.listdir(data_directory):
    class_dir = os.path.join(data_directory, class_name)
    # 만약 현재 항목이 디렉토리라면 진행
    if os.path.isdir(class_dir):
        # 현재 클래스(폴더) 내의 각 이미지 파일을 반복
        for filename in os.listdir(class_dir):
            # 파일명이 확장자가 .png인지 확인하고 이미지 파일인지 확인
            if filename.lower().endswith('.png') and os.path.isfile(os.path.join(class_dir, filename)):
                # 이미지 파일 경로를 생성
                img_path = os.path.join(class_dir, filename)
                # 이미지 파일 열기
                img = Image.open(img_path)
                # 이미지 크기 조정 및 RGB로 변환하여 리스트에 추가
                images.append(np.array(img.resize((224, 224)).convert('RGB')))
                # 클래스명(폴더명)을 레이블 리스트에 추가
                labels.append(class_name)

# numpy 배열로 변환
X = np.array(images)
y = np.array(labels)

# 결과 확인
print("데이터 shape:", X.shape)
print("레이블 shape:", y.shape)
```

```python
from PIL import Image
import os

def convert_pngs_to_jpgs(input_folder, output_folder):
    # 출력 폴더가 없으면 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 입력 폴더에서 PNG 파일 목록 가져오기
    png_files = [f for f in os.listdir(input_folder) if f.endswith('.png')]

    for png_file in png_files:
        try:
            # PNG 파일 열기
            png_path = os.path.join(input_folder, png_file)
            png_image = Image.open(png_path)

            # JPG 파일로 변환하여 저장
            jpg_path = os.path.join(output_folder, os.path.splitext(png_file)[0] + '.jpg')
            jpg_image = png_image.convert("RGB")
            jpg_image.save(jpg_path)

            print("변환 완료:", jpg_path)
        except Exception as e:
            print("오류 발생:", e)

# 변환할 PNG 파일이 있는 폴더 경로 설정
input_folder = "/content/drive/MyDrive/deeplearning/remocon/ssm-original4"

# JPG 파일이 저장될 폴더 경로 설정
output_folder = "/content/drive/MyDrive/deeplearning/remocon/ssm-original5"

# PNG를 JPG로 변환
convert_pngs_to_jpgs(input_folder, output_folder)
```

```python
import os

def change_extensions_to_jpg(folder_path):
    # 폴더 안에 있는 모든 파일 목록 가져오기
    files = os.listdir(folder_path)

    for file_name in files:
        try:
            # 파일의 전체 경로 구성
            file_path = os.path.join(folder_path, file_name)

            # 파일의 확장자 추출
            _, file_extension = os.path.splitext(file_name)

            # 파일의 확장자가 jpg가 아닌 경우에만 변경
            if file_extension.lower() != '.jpg':
                # 새 파일 이름 구성 (확장자를 jpg로 변경)
                new_file_name = file_name + '.jpg'
                new_file_path = os.path.join(folder_path, new_file_name)

                # 파일 이름 변경
                os.rename(file_path, new_file_path)

                print("변환 완료:", new_file_path)
            else:
                print("이미 jpg 확장자입니다:", file_path)
        except Exception as e:
            print("오류 발생:", e)

# 확장자를 변경할 폴더 경로 설정
folder_path = "/content/drive/MyDrive/deeplearning/remocon/ssm-print"

# 파일 확장자를 jpg로 변경
change_extensions_to_jpg(folder_path)
```

test_size = int(len(data)*.1): 전체 데이터셋의 10%를 테스트 세트로 사용하기 위해 전체 길이에 0.1(10%)를 곱한 후, 정수로 변환하여 테스트 세트의 크기를 계산합니다.

이렇게 학습, 검증, 테스트 세트의 크기를 계산하면 데이터셋을 적절히 나눌 수 있습니다. 일반적으로 학습 세트의 크기가 가장 크고, 검증 및 테스트 세트의 크기는 작게 설정됩니다.

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```


train_size 변수에 할당된 값이 11이라는 것은 학습 세트의 크기가 11이라는 의미입니다. 이것은 데이터셋을 학습, 검증, 테스트 세트로 나눌 때 사용된 비율에 따라 결정됩니다. 보통 데이터셋을 70%의 학습 세트, 20%의 검증 세트, 그리고 10%의 테스트 세트로 나눕니다. 따라서 전체 데이터셋에서 학습 세트의 크기가 약간 작게 설정되었을 경우, 학습 세트의 크기가 11인 것입니다.


해당 코드는 데이터셋을 학습, 검증, 테스트 세트로 나누는 작업을 수행합니다. 아래는 코드의 해석입니다.

train = data.take(train_size): 전체 데이터셋(data)에서 학습 세트의 크기(train_size)만큼의 샘플을 가져와서 학습 세트(train)로 지정합니다. take() 메서드는 데이터셋에서 주어진 수만큼의 요소를 가져오는 데 사용됩니다. 따라서 data.take(train_size)는 처음부터 학습 세트 크기(train_size)까지의 샘플을 가져오는 것입니다.

val = data.skip(train_size).take(val_size): 전체 데이터셋(data)에서 학습 세트의 크기만큼의 샘플을 건너뛴 다음, 검증 세트의 크기(val_size)만큼의 샘플을 가져와서 검증 세트(val)로 지정합니다. skip() 메서드는 주어진 수만큼의 요소를 건너뛰는 데 사용됩니다. 따라서 data.skip(train_size).take(val_size)는 학습 세트 다음부터 검증 세트 크기(val_size)만큼의 샘플을 가져오는 것입니다.

test = data.skip(train_size+val_size).take(test_size): 전체 데이터셋(data)에서 학습 세트와 검증 세트의 크기를 합친 만큼의 샘플을 건너뛴 다음, 테스트 세트의 크기(test_size)만큼의 샘플을 가져와서 테스트 세트(test)로 지정합니다. 이는 이전과 같이 skip()과 take() 메서드를 조합하여 수행됩니다.

따라서 위 코드는 전체 데이터셋을 학습, 검증, 테스트 세트로 나누는 과정을 수행합니다. 이를 통해 각각의 세트에 적절한 샘플이 할당됩니다.

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

Conv2D(16, (3,3), 1, activation='relu', padding='same', input_shape=(256,256,3)):

2D 컨볼루션 레이어를 추가합니다.
16개의 필터를 사용합니다.
필터의 크기는 (3,3)입니다.
스트라이드(stride)는 1로 설정합니다.
활성화 함수(activation)로 ReLU를 사용합니다.
패딩(padding)은 'same'으로 설정하여 입력과 출력의 크기를 동일하게 유지합니다.
입력 이미지의 모양은 (256,256,3)으로 지정합니다.
MaxPooling2D(pool_size=(2,2), strides=(2,2)):

2D 맥스 풀링 레이어를 추가합니다.
풀링 크기(pool_size)는 (2,2)로 설정합니다.
스트라이드(strides)는 (2,2)로 설정하여 풀링을 수행할 때마다 입력 이미지의 크기를 반으로 줄입니다.
위와 같은 방식으로 Conv2D와 MaxPooling2D 레이어를 여러 번 중첩하여 신경망을 깊게 구성합니다. 주석 처리된 부분은 필요에 따라 레이어를 추가할 수 있습니다.

Flatten(): Flatten 레이어는 2D 배열을 1D 배열로 변환합니다. 이는 컨볼루션 레이어와 맥스 풀링 레이어의 출력을 완전 연결 레이어에 연결하기 위해 사용됩니다.

Dense(256, activation='relu'):

완전 연결 레이어를 추가합니다.
256개의 뉴런을 가집니다.
활성화 함수로 ReLU를 사용합니다.
Dense(1, activation='sigmoid'):

출력 레이어로서, 이진 분류 문제를 다루는 경우에 사용됩니다.
1개의 뉴런을 가집니다.
활성화 함수로 Sigmoid를 사용하여 출력을 [0, 1] 범위로 변환하여 이진 분류 확률을 예측합니다.
이렇게 구성된 신경망은 Convolutional Neural Network(CNN) 구조를 따르며, 이미지 분류나 이진 분류와 같은 작업에 사용될 수 있습니다.

model.compile(): 모델을 컴파일하는 메서드입니다. 이 메서드를 사용하여 모델의 학습 과정을 설정합니다.

loss=tf.losses.BinaryCrossentropy(): 손실 함수(loss function)로서 이진 분류에 사용되는 Binary Crossentropy를 사용합니다. Binary Crossentropy는 이진 분류 문제에서 확률 분포 간의 차이를 측정하는 손실 함수로, 모델의 출력과 실제 타겟 간의 오차를 계산합니다.

metrics=['accuracy']: 모델의 성능을 평가하기 위한 지표(metrics)로서 정확도(accuracy)를 사용합니다. 정확도는 모델이 올바르게 분류한 샘플의 비율을 나타내는 지표로, 학습 과정에서 모니터링되고 최적화에 사용됩니다.

위 과정을 두 번 반복하고 있습니다. 이는 컨볼루션과 풀링을 두 번 적용하여 이미지의 특징을 추출하고 중요한 정보를 강조하는 과정을 반복함으로써 모델이 더욱 복잡한 패턴을 학습할 수 있도록 합니다.

model.add(Flatten()): 2차원 특징 맵을 1차원으로 펼치는 과정을 수행합니다. 이는 컨볼루션과 풀링 과정을 거친 후의 특징 맵을 1차원으로 변환하여 완전 연결 계층에 입력으로 제공하기 위한 과정입니다.

model.add(Dense(256, activation='relu')): 256개의 뉴런으로 구성된 완전 연결(Dense) 계층을 추가합니다. 활성화 함수로는 ReLU를 사용합니다.

model.add(Dense(1, activation='sigmoid')): 이진 분류 문제를 다루는 마지막 출력 계층을 추가합니다. 활성화 함수로는 시그모이드(Sigmoid) 함수를 사용하여 출력을 0과 1 사이의 값으로 제한합니다. 이는 출력이 이진 분류 확률을 나타내도록 합니다.

tf.keras.callbacks.TensorBoard(log_dir=logdir): 텐서보드를 사용하여 모델의 학습 과정을 시각화하기 위한 콜백을 생성합니다. 이 콜백은 학습 중에 모델의 성능 지표 및 그래프, 히스토그램 등을 텐서보드에 기록합니다.

따라서 위 코드는 모델의 학습 과정을 시각화하기 위한 텐서보드 콜백을 생성하는 것입니다. 이 콜백은 학습 중에 지정된 디렉토리에 로그를 저장하여 텐서보드가 학습 과정을 시각적으로 표현할 수 있도록 합니다.

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

validation_data=val은 검증 데이터셋을 나타냅니다. 학습 중에 검증 데이터셋을 사용하여 모델의 성능을 평가합니다.
callbacks=[tensorboard_callback]은 텐서보드 콜백을 모델에 적용하여 학습 과정 중 로그를 저장하도록 설정합니다. 이를 통해 학습 과정을 시각화할 수 있습니다.
hist에는 학습 과정에서의 손실(loss) 및 지정된 지표(metric)의 값들이 기록됩니다.
print(hist.history): hist.history는 학습 과정에서 기록된 지표들을 담고 있는 딕셔너리 객체를 출력합니다. 딕셔너리의 키는 각 지표를 나타내며, 해당 지표의 값들이 리스트 형태로 저장되어 있습니다.

best_score = max(hist.history['val_accuracy']):

학습 과정에서 기록된 검증 세트에 대한 정확도('val_accuracy') 중에서 최고 성능을 나타내는 값을 best_score 변수에 저장합니다.

print(f"Best Validation score is: {best_score}"): 최고 검증 세트 정확도를 출력합니다. 이 값은 모델이 학습 과정 중에 가장 좋은 성능을 보였을 때의 정확도를 나타냅니다.

모델성능평가, 정확도, 정밀도, 재현율, f1-score, 혼동행렬

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 모델 예측
# y_pred = model.predict(X_test)
y_pred = model.predict(X)
# 이진 분류일 경우 threshold를 지정하여 0 또는 1로 변환
# 예를 들어, 임계값이 0.5일 때, 0.5 이상의 값은 1로, 0.5 미만의 값은 0으로 변환
y_pred_binary = (y_pred > 0.5).astype(int)

# 정확도(Accuracy) 계산
# accuracy = accuracy_score(y_test, y_pred_binary)
accuracy = accuracy_score(y, y_pred_binary)

# 정밀도(Precision) 계산
precision = precision_score(y, y_pred_binary)

# 재현율(Recall) 계산
recall = recall_score(y, y_pred_binary)

# F1-score 계산
f1 = f1_score(y, y_pred_binary)

# 혼동 행렬(Confusion Matrix) 계산
conf_matrix = confusion_matrix(y, y_pred_binary)

# 결과 출력
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```
